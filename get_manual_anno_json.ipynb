```python
from PIL import Image, ImageDraw, ImageFont
import json
import os
with open('/content/drive/MyDrive/OCR_data_all.json') as f:
  json_all = json.load(f)
```

```python
f = open("/content/drive/MyDrive/annotation.txt",'r')
file_names_with_page = f.read().splitlines()
file_names = []
for file_name in file_names_with_page:
  name = file_name.split('-')[0]
  if not name in file_names:
    file_names.append(name)
len(file_names)
```

```python
import json
base_path = "/content/drive/MyDrive"
def save_json(dict_data):
  try:
    json_file = open(base_path + '/manual_anno.json', "w")
    json.dump(dict_data, json_file)
    print("Success saved!")
  except:
    print("    save_json(): fail save as JSON")
```

```python
new_json = {}
for file_name in file_names:
  json_temp = json_all[file_name + '.pdf']
  for page_id in range(len(json_temp['pages'])):
    page_id = str(page_id)
    for object_id in range(len(json_temp['pages'][page_id]['objects'])):
      object_id = str(object_id)
      for sub_object_id in range(len(json_temp['pages'][page_id]['objects'][object_id]['sub_obj'])):
        sub_object_id = str(sub_object_id)
        json_temp['pages'][page_id]['objects'][object_id]['sub_obj'][sub_object_id]['label'] = 'text'
  new_json[file_name + '.pdf'] = json_temp
save_json(new_json)
```
